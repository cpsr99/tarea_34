```python
# ejemplo - generando una operacion a ser aplicada a un listado de numeros 
# este ejemplo es SIN MAP()
# generando un listado de numeros con exponencial 

# declara las listas a ser usadas
org_list = [1, 2, 3, 4, 5]
fin_list = []

# aplica la operacion a cada uno de los numeros en la lista
for num in org_list:
    fin_list.append(num**3) # ** = potenciacion (el 3 lo hace al cubo) # o sea que los numero que estaran en fin_list son los numeros org_list al cubo

# imprime la lista
print(fin_list)
```

```python
# mismo ejemplo exponencial usando map()

# define la lista inicial de numeros, no es necesario definir la lista resultado
org_list = [1, 2, 3, 4, 5]

# define la uncion a ser aplicada a cada numero 
def cube(num):
    return num**3

# define la funcion map()
fin_list = list(map(cube, org_list))
print(fin_list)

# estructura de cogigo mucho mas limpia y rapida 
# se pueden aplicar a cualquier numero de funciones y ademas aclarar las funciones en otros archivos incluso 
```

```python
# otro ejemplo
# largo de cada elemento en una vista 
# utilizo la funcion de python "len" - cuenta el largo de cada palabras e imprime elnumero de letras 
org_list = ["Hey", "mundo", "EBAC"]
fin_list = list(map(len, org_list))
print(fin_list)
```

```python
# usando mas de un iterable (base y power)
# aprovechando la funcion pow de python
base = [10, 10, 10, 10]
power = [1, 2, 3, 4]

result = list(map(pow, base, power))
print(result)
```

```python
# ejemplo map + lambdas = codigo mucho mas limpio 
org_list = [1,2,3,4,5]

fin_list = list(map(lambda x:x**3, org_list))
print(fin_list) # 1,8,27,64,125

# se puuede reducir aun mas 
org_list = [1,2,3,4,5]

print(list(map(lambda x:x**3, org_list)))
```

```python
# doble de un numero 
def addition(n):
    return n + n
# duplicamos todos los numero usando map
numbers = (1,2,3,4)
result = map(addition, numbers)
print(list(result))
```

```python
# dobla los numeros usando map y lambda 
numbers = (1,2,3,4)
result = map(lambda x: x + x, numbers)
print(list(result))

# simplfica el ejemplo anterior
```

```python
# suma de dos listas con map y lambda 
numbers1 = [1,2,3]
numbers2 = [4,5,6]
numbers3 = [7,8,9]

result = map(lambda x, y, z: x + y + z, numbers1, numbers2, numbers3)
print(list(result))
```

```python
# ejemplo con cadenas de caracteres
l = ['sat', 'bat', 'cat', 'mat']

# map() puede usarse para dividir en caracteres a las palabras enviadas
# obteniendo una lista de listas 
test = list(map(list, l)) # list divide por caracter, "deletrea"
print(test)
```
