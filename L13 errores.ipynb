# 4.- error de tipo excepción

```python
# generando excepciones en el código
# no son errores de python, sino condiciones que van a cortar el flujo del programa

x = 8
if x > 7: 
    raise Exception('x no puede sobrepasar el valor 7. El valor de x fue: {}'.format(x))
print(x)

# se podria decir que es un error que yo misma estoy creando
```

# 4.1 Evaluación de condiciones con Assert 
si se cumple la condición, se genera la excepción

```python
x = "hola"

# no ocurre nada si la condición se cumple
#assert x == "hola"

# si la condición es falsa, se genera una excepción
assert x == "ciao" 
```

```python
# tambien se pueden evaluar las condiciones (assert) que causan el problema
# en caso de que no se cumplan, las deja fuera 

# importa la libreria "sys" que validara que tipo de plataforma tenemos 
import sys
print('Platform= ', sys.platform)
assert ('win32' in sys.platform), "Este código nada más funciona en windows" # este mensaje saldra en la línea de error
```

```python
# mismo ejemplo con diferente plataforma 
import sys
print ('Platform= ', sys.platform)
assert ('darwin' in sys.platform), "Este código nada más funciona en mac" # este mensaje saldra en la línea de error
```

```python
# la funcion assert se usa para validar condiciones tanto de calculo, de negocio, de sistema, etc
# corta el flujo del sistema cuando se cree necesario, antes de pasar a ejecuciones más largas 
```

```python
# ejemplo 1 - ambiente de desarrollo
def linux_interaction():
    assert ('linux' in sys.platform), "Solo linux"
    print('Trabajando con linux')

try: 
    linux_interaction()
except:
    print('No linux')
    pass
```

```python
# se modifica ahora el código del except con un "assert"
try:
    linux_interaction()
except AssertionError as error:
    print(error)
    print('La función linux_interaction() no se ejecutó')
```

```python
# ejemplo con else
# se imprime el reciproco de un numero 
# ademas se utiliza la funcion "input" para utilizar la barra de comandos para ingresar el numero 
# si es par, genera el reciproco, si es impar, genera mensaje

try:
    num = int(input("Ingrese un número: "))
    assert num % 2 == 0
except: 
    print("No ingresaste un número par")
else: 
    reciprocal = 1/num
    print(reciprocal)
```

```python
# ejemplo 3 - division desde 0

def divide(x, y):
    try:
        # se utiliza la division Floor, solamente fraccional
        result = x // y
        print ("Respuesta: ", result)
    except ZeroDivisionError:
        print("División para cero")

divide(7, 2) # respuesta 1
divide(7, 0) # respuesta 2

```

```python
# modificacion menor de ejemplo anterior 
def divide(x, y):
    try:
        result = x // y
    except ZeroDivisionError:
        print("División para cero")
    else: 
        print ("Respuesta: ", result)

divide(10, 2) # respuesta 1
divide(10, 0) # respuesta 2
```

# FINALLY

```python
def divide(x, y):
    try:
        result = x // y
    except ZeroDivisionError:
        print("División para cero")
    else: 
        print("Respuesta: ", result)
    finally: # este bloque se ejecuta SIEMPRE, sin importar el estado de las excepciones
        print('Fin del ejercicio')

divide(3, 2)
divide(3, 0)
```
